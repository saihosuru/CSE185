```python
# This isolates the reads from chromosome 17 from BWA-MEM SAM file 
reads_chr17 = []

with open("Klf4.esc.sam", "r") as sam_file:
    for line in sam_file:
        if line.startswith("@"):  # Skip header lines
            continue
        cols = line.split("\t")
        chromosome = cols[2]
        if chromosome == "17":
            reads_chr17.append(line)
```

```python
# We print out how the reads are being stored 
i = 0

while i < 40:
    print(reads_chr17[i])
    i += 1
```

```python
# We isoalate chromosome 17 reads with the SAM file we generated 
import zipfile
import pandas as pd

zip_name = "large_result.zip"
reads_fromourfile = []

with zipfile.ZipFile(zip_name,'r') as zip_file:
    our_result = zip_file.namelist()[0]
    with zip_file.open(our_result) as file:
        for line in file:
            line = line.decode("utf-8")
            fields = line.split("\t")
            chromosome = fields[3]
            if chromosome == "17":
                reads_fromourfile.append(line)
        
```

```python
# We print the reads out to see how they are being stored 

i = 0

while i < 40:
    print(reads_fromourfile[i])
    i += 1
```

```python
print(len(reads_chr17))

print(len(reads_fromourfile))


# We see how many reads are in chromsome 17 for the different sam files 
```

```python
# NOW WE CREATED A DICTIONARY WHERE THE KEY IS THE READ ID AND THE VALUE IS THE CHROMOSOME NUMBER AND GENOMIC POSITION AND WE 
# GO THROUGH THE READS IN THE SAME FILE WE GENRATE AND CHECK TO SEE IF THERE ARE ANY MATCHES 

reads_chr17_dict = {}
reads_fromourfile_dict = {}
Matching = []
read_id = []
for elem in reads_chr17:
    cols = elem.split("\t")
    if cols[0].strip() not in reads_chr17_dict:
        reads_chr17_dict[cols[0]] = [(cols[2],cols[3])]
    elif cols[0].strip() in reads_chr17_dict:
        reads_chr17_dict[cols[0].strip()].append((cols[2],cols[3]))
      #3,4  
for elem in reads_fromourfile:
    cols = elem.split("\t")
    cols[0] = cols[0].replace("@","")
    cols[0] = cols[0].strip()
    if cols[0] in reads_chr17_dict:
        cols[4] = cols[4].replace("\n","")
        read_id.append(  [  (cols[0],cols[4]),  reads_chr17_dict[cols[0]] ])
        for tup in reads_chr17_dict[cols[0]]:
            if (cols[3] == tup[0] and (int(cols[4]) == int(tup[1]) or int(cols[4]) + 1 == int(tup[1]))):
                Matching.append(cols)
```

```python
# We print out the matched reads to see how they are stored 
i = 0

while i < 40:
    print(Matching[i])
    i += 1
    
print(len(Matching))
```
